```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
import seaborn as sns

```

```python
df =datasets.load_iris()
df = pd.DataFrame(data=df.data, columns=df.feature_names)
df
```

```python
sns.pairplot(df)
```

```python
inercia = []
for i in range(2,7):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    inercia.append(kmeans.inertia_)
print(inercia)
```

```python
plt.plot(range(2,7),inercia,marker='o')

#plt.yscale('log')
plt.title('Curva Elbow')
plt.ylabel('Inercia')
plt.xlabel('Clusters')
plt.show()
```

```python
kmeans_final = KMeans(n_clusters = 3)
kmeans_final.fit(df)
```

```python
clusters = kmeans_final.cluster_centers_
clusters
```

```python
pred = kmeans_final.predict(df)
pred
```

```python
plt.scatter(df[pred==0]['sepal length (cm)'],df[pred==0]['sepal width (cm)'],label='cluster1')
plt.scatter(df[pred==1]['sepal length (cm)'],df[pred==1]['sepal width (cm)'],label='cluster2')
plt.scatter(df[pred==2]['sepal length (cm)'],df[pred==2]['sepal width (cm)'],label='cluster2')
plt.show()

```

```python
plt.scatter(df[pred==0]['sepal length (cm)'],df[pred==0]['petal length (cm)'],label='cluster1')
plt.scatter(df[pred==1]['sepal length (cm)'],df[pred==1]['petal length (cm)'],label='cluster2')
plt.scatter(df[pred==2]['sepal length (cm)'],df[pred==2]['petal length (cm)'],label='cluster2')
plt.show()
```
